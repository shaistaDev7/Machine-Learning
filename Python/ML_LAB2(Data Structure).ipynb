# **What Are Data Structures?**

Data structures are special containers that **store multiple** values in Python.
They help us:



*    Organize data


*   Modify and manage large amounts of information easily



*    Access data faster



There are 4 important data structures in Python:


*   **Lists**

*   **Tuples**
*   **Dictionaries**

*   **Sets**








 # **1. Lists**

A list is an **ordered collection** of items.
It is **changeable**, which means we can add, remove, or update values.

**✔ Characteristics**


*   Written in square brackets → [ ]
*   Items are indexed (starting from 0)


*   Supports adding/removing items



```python
fruits = ["apple", "banana", "mango"]
marks = [85, 90, 78]
mix=["Apple", 90, 78.5] #In list we can store mutiple types of data
print(fruits)
print(marks)
print(mix)
```

# **List Operations**
 **1. Accessing elements**


```python
fruits = ["apple", "banana", "mango"]
print(fruits[2]) #use indexing to access element from list
```

**2. Changing element**

```python
fruits[1] = "orange"
print(fruits)  # ["apple", "orange", "mango"]
```

**3. Adding items**

```python
fruits.append("grapes") # append() add an element at the end of the list.
print(fruits)
```

```python
fruits.insert(1,"banana") #insert() Adds an element at a specific position.
print(fruits)
```

**4. Removing items**
* Remove an element by value
* Remove an element using index

```python
#remove an element by value
fruits.remove("apple") #remove() is used to remove a specific value.
print(fruits)
```

```python
#remove an element by index
fruits.pop(2) #pop() is used to remove an element by index position.
print(fruits)
```

```python
fruits.pop() #If positon is not specify it removes last ement 
print(fruits)
```

**5. List length**

```python
print(len(fruits))
```

**6. Index()**

```python
print(fruits.index("banana")) #index() is used to find the position of elment 
```

**7.  sort()**

```python
numbers = [5, 2, 8, 1]
numbers.sort() #sort elements in ascending order
print(numbers)
```

```python
numbers = [5, 2, 8, 1]
numbers.sort(reverse=True) #sort elements in descending order
print(numbers)
```

# **2. Tuples**

A tuple is similar to a list but **cannot be changed**.
It is immutable.

**✔ Characteristics**



*   Written in parentheses → ( )

*   Ordered
*   Cannot modify values





**Accessing tuple elements**

```python
colors = ("red", "green", "blue")
print(colors[0])  # red

thistuple = ("apple", "banana", "cherry")
print(thistuple[-1]) #cherry #-1 means index start from last
```

**Add Item**

Tuples are immutable, which means we cannot directly add, remove, or update values after the tuple is created But if you want to add item in tuple then you should first convert tuple into list->add the item->convert back into a tuple.

```python
thistuple = ("apple", "banana", "cherry")
y = list(thistuple) #convert typle into list
y.append("orange")  #add item into list
thistuple = tuple(y) #convert back list into tuple
print(thistuple)
```

**Update tuple**

```python
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
```

**Tuples cannot be changed**

```python
colors[1] = "yellow"
```

# **3. Dictionaries**

Dictionaries store data in key : value pairs.
Each item has a name (key) and a value.

**✔ Characteristics**



*   Data is stored as key-value pairs






*   Written in curly braces → { }




*   Keys must be unique
*   Values can be changed



```python
student = {
    "name": "Sara",
    "age": 20,
    "grade": "A"
}
```

**Dictionary Operations**

**Accessing value**

```python
student = {"name": "Sara", "age": 20}
print(student["name"]) #we can access dictionary value using key
```

**Changing values**

```python
student["age"] = 21
print(student)
```

**Adding new key-value**

```python
student["city"] = "Sialkot"
print(student)
```

**Removing key-value**

```python
student.pop("age") #it will remove key-value pair
print(student)
```

**keys()**

```python
print(student.keys()) #it returns all keys from dictionary
```

**values()**

```python
print(student.values()) #it returns all values from dictionary
```

**items()**

```python
print(student.items()) #it returns keys and values both from dictionary
```

# **4. Sets**

A set in Python is a **collection of unique items**.

It is mostly used when you want to remove duplicates  or check membership quickly.

**✔ Characteristics of Sets:**

*   Written in curly braces → { }

*   No duplicates allowed
*   Unordered (items have no index)


*   Changeable (you can add/remove items)

```python
duplicates = {1, 2, 2, 3}
print(duplicates)
# Output: {1, 2, 3}
# Python automatically removes duplicate "2"
```

**Set Operations**

**1. Creating a Set**

```python
my_set = {10, 20, 30}
print(my_set)
```

**2. Adding Items**

```python
fruits = {"apple", "banana"}
fruits.add("mango")
print(fruits)
```

 **3. Removing Items**

```python
fruits.remove("banana")
print(fruits)
```

**4. Checking If Value Exists**

```python
numbers = {5, 10, 15, 20}
print(10 in numbers)   # True
print(50 in numbers)   # False
```

**5. Set Length**

```python
print(len(numbers))
```

**6. Removing Duplicates Using Set**

```python
data = [1, 2, 2, 3, 3, 3, 4]
unique_data = set(data)
print(unique_data)  # {1, 2, 3, 4}
```

**7. Union (Combine Sets)**

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
# Output: {1, 2, 3, 4, 5}
```

**8. Intersection (Common Values)**

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
# Output: {2, 3}
```

**9. Difference**

```python
A = {1, 2, 3}
B = {2, 3, 4}

print(A.difference(B))
```

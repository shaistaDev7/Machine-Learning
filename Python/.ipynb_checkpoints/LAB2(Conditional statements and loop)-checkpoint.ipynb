# **Conditional statements**

 Help your program to make decisions.



**Main Types of Conditional Statements**


if statement

if-else statement

if-elif-else chain



### if statement

```python
temperature = int(input("Enter temperature"))
if temperature>30:
    print("Today is hot")
```

### if-else statement

**Write a Python program that takes marks as input and prints:**

* "Pass" if marks are 40 or above

* "Fail" otherwise

```python
marks = int(input("Enter your marks: "))

if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

### Using if-elif-else (Multiple Conditions)

**Write a Python program that checks whether a number is**:

* Positive

* Negative

* Zero

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

**Write a Python program that prints grades based on marks:**

* A Grade (≥ 80)

* B Grade (≥ 60)

* C Grade (≥ 40)

* Fail (below 40)

```python
marks = int(input("Enter marks: "))

if marks >= 80:
    print("A Grade")
elif marks >= 60:
    print("B Grade")
elif marks >= 40:
    print("C Grade")
else:
    print("Fail")
```

# **For Loop**

**What is a Loop?**



A loop l**ets you repeat a block of code** as many times as needed — just like when you say:

 “Keep eating until the plate is empty.”

**Using for with range()
Perfect for repeating a task multiple times:**

A for loop is commonly used when the number of iterations is known in advance.

```python
print("Counting to 5...")

for i in range(6):
    print(i, end=" ")
print("\nDone!")
```

```python
for i in range(4, 9): #include, exclude
    print(i)
```

```python
for i in range(0, 11, 2): #0 starting, 10 ending exclude, 2 skip, 2 gap 0,
    print(i)
```

```python
fruits = ["apple", "banana", "cherry"] 
for x in fruits: # for fruit in fruits
    print(x)
```

```python
fruits = ["apple", "banana", "cherry"] 
for x in range(len(fruits)): 
    print(fruits[x])
```

# **While Loop**
A while loop repeatedly executes a block of code as long as a condition remains True. It is useful when the number of iterations is not known beforehand.

**Basic while**

```python
i = 0 #1, 5
while i < 5:
    print(i)
    i += 1
```

```python
is_running = True
count = 0

while is_running: #true
    print(count)
    count += 1 #0 , 3
    if count == 4:
        is_running = False
```

## Loop Flow Control

Loop flow control statements change the normal execution of loops. Python provides break and continue control loop behavior.

**Using break**

```python
for i in range(10):
    if i == 5:
        break #end here
    print(i)
```

**Using continue**

```python
for i in range(5):
    if i == 2:
        continue #2 ,start the loop again
    print(i)
```

```python
import random

secret = random.randint(1, 20) #take random number between 1 and 20 every time when we run this code
print(secret)
```

```python
import random

secret = random.randint(1, 20)
guess = None
attempts = 0

print("Guess the secret number between 1 and 20!")

while guess != secret:
    guess = int(input("Your guess: "))
    attempts += 1
    if guess < secret:
        print("Too low! Try again")
    elif guess > secret:
        print("Too high! Try again")

print(f"You got it in {attempts} attempts!")
```

**Nested Loops – Loops Inside Loops**



Great for tables, grids, or patterns. Example: Multiplication table, pattern printing:

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i*j}", end=" ")
    print()  # new line after inner loop
```

#**File Handling**

In Python we can **open, read, write,** and **close files** stored on the computer.

**Why Do We Use File Handling?**

To store data permanently.

**Example:**

When you close a Python program, variables disappear

BUT if you write data into a file, it stays saved

You can open the file again anytime

**Open the file**

```python
file = open("data.txt", "r")  # open file in read mode
content=file.read()
print(content)
```

**Close the file**

```python
file.close()
```

**Reading the file**

```python
file = open("data.txt", "r")  # open file in read mode
content=file.read()
print(content)
file.close()
```

**Writing to a file**

```python
file = open("data.txt", "w")
file.write("Hello IT CS6")
file.close()
```

```python
file = open("data.txt", "r")  # open file in read mode
content=file.read()
print(content)
```

**Appending to a file**

```python
file = open("data.txt", "a")
file.write("\ntoday we are learning file handling concept")
file.close()
```

```python
file = open("data.txt", "r")  # open file in read mode
content=file.read()
print(content)
```

**readlines() – Gives list of all lines**

```python
file = open("data.txt", "r")
print(file.readlines())
file.close()
```

```python
#Read only second line just then
file = open("data.txt", "r")
result=file.readlines()
print(result[1])
file.close()
```

**Reading line by line**

```python
file = open("data.txt", "r")
for line in file:
    print(line)
file.close()
```

**readline() – Reads only ONE line**

```python
file = open("data.txt", "r")
print(file.readline())
print(file.readline())
file.close()
```

**Using "with"**

**"Always close the file after using it."**

OR use:

**with keyword**(auto close)

```python
with open("data.txt","r") as file:
  print(file.read())
```

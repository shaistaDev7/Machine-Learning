# PANDAS
Pandas is a powerful open-source Python library used for data analysis and data manipulation.
## Why Pandas is used?

It helps:

- Work with large datasets
- Handle missing data
- Perform data cleaning
- Analyze data quickly
- Read/write data from files (CSV, Excel, etc.)


```python
# install and import library
import numpy as np
import pandas as pd
```

## Main Data Structures in Pandas

### 1. Series
- One-dimensional (like a single column)
- Example: list of marks

### 2. DataFrame
- Two-dimensional (like a table with rows and columns)
- Example: student records table

### Creating DataFrame

```python
# using lists
student_data = [
    [100,80,90],
    [90,70,70],
    [80,100,95],
    [80,50,70]
]

pd.DataFrame(student_data, columns=['English','Science','Math'], index=['a','b','c','d'])
```

```python
# using dicts

student_dict = {
    'Name':['Sara','Ayesha','Rabia','Arooj'],
    'English':[100,90,80,80],
    'Science':[80,70,100,50],
    'Math':[90,70,95, 70]
}

students_rec = pd.DataFrame(student_dict)
students_rec
```
